```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_rewards(all_rewards_1, all_rewards_2, name, threshold=475):
    """
    Plots the rewards for Mean Dueling DQN and Max Dueling DQN over episodes.
    
    Parameters:
    all_rewards_1 (list): List of rewards for Mean Dueling DQN.
    all_rewards_2 (list): List of rewards for Max Dueling DQN.
    name (str): Name of the environment.
    
    Returns:
    None
    """
    
    std_dev_1 = np.std(all_rewards_1)
    std_dev_2 = np.std(all_rewards_2)
    x_1 = np.arange(0,len(all_rewards_1))
    x_2 = np.arange(0,len(all_rewards_2))
    plt.fill_between
    plt.xlabel('Episode')
    plt.ylabel('Total Reward')
    plt.plot(x_1, all_rewards_1, label='Mean Dueling DQN (Type-1)',color = 'blue')
    plt.plot(x_2, all_rewards_2, label='Max Dueling DQN (Type-2)', color = 'red')
    plt.fill_between(x_1,
                     np.array(all_rewards_1) - std_dev_1/2,
                     np.array(all_rewards_1) + std_dev_1/2,
                     color='blue', alpha=0.2, label='± 0.5 SD')
    plt.fill_between(x_2,
                     np.array(all_rewards_2) - std_dev_2/2,
                     np.array(all_rewards_2) + std_dev_2/2,
                     color='red', alpha=0.2, label='± 0.5 SD')
    threshold_array = np.full((len(all_rewards_1),), threshold)
    plt.plot(threshold_array, 'g--', label='Threshold', color = 'black')
    plt.title(f'Mean Dueling DQN vs Max Dueling DQN for {name}')
    plt.legend()
    plt.show()
```

```python
all_rewards_1 = np.load('Results_DQN/mean_cartpole.npy')
all_rewards_2 = np.load('Results_DQN/max_cartpole.npy')
plot_rewards(all_rewards_1, all_rewards_2, 'CartPole-v1', 475)
```

```python
all_rewards_1 = np.load('Results_DQN/mean_acrobot.npy')
all_rewards_2 = np.load('Results_DQN/max_acrobot.npy')
plot_rewards(all_rewards_1, all_rewards_2, 'Acrobot-v1', -100)
```
